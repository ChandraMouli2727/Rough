```python

import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python

dir(digits)
```

```python

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
df[0:12]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
df.drop(['target'],axis='columns'),digits.target,test_size = 0.2)
```

```python
len(X_test),len(y_test)
```

```python
# ensemble when we used multiple algorithm to predict outcome
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 120)
model.fit(X_train,y_train) # n_estimators it is used 100 random trees if u increase score is increasing
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
df.drop(['target'],axis='columns'),iris.target,test_size = 0.2)
```

```python
len(X_train),len(y_train)
```

```python
# ensemble when we used multiple algorithm to predict outcome
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train) # n_estimators it is used 100 random trees if u increase score is increasing
```

```python
model.score(X_test,y_test)
```

```python
# ensemble when we used multiple algorithm to predict outcome
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 40)
model.fit(X_train,y_train) # n_estimators it is used 100 random trees if u increase score is increasing
```

```python
model.score(X_test,y_test)
```
