* It is Used to reduce Dimension
* what if we get rid of non important features ?
* 1.Faster training and inference
* 2.Data Visualization becomes easier
* PCA is a process of figuring out most important features or principal components that has the most impact on the target variable
* scale features before applying
* accuracy might drop
* PCA is called dimensionality reduction technique as it can help us reduce dimensions
* Dimensionality curse

```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
ds = load_digits()
ds.keys()
```

```python
ds.data.shape
```

```python
ds.data[0] # it is just numpy array
```

```python
ds.data[0].reshape(8,8)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.gray()
plt.matshow(ds.data[9].reshape(8,8))
```

```python
import numpy as np
np.unique(ds.target)
```

```python
ds.target[9]
```

```python
df = pd.DataFrame(ds.data,columns=ds.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y=ds.target
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size =0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95) #Retain 95 % useful features
X_pca = pca.fit_transform(X)
X_pca.shape # reduced columns using pca technique
```

```python
X.shape[1]-X_pca.shape[1] # check how many columns get rid off
```

```python
pca.explained_variance_ratio_ # each col how much information consists in %
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca,y,test_size =0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca) # Accuracy is similiar but cols reduced
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca,y,test_size =0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca) # Accuracy is similiar but cols reduced
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python

dfex[dfex.RestingBP>(dfex.RestingBP.mean()+3*dfex.RestingBP.std())]
```

```python

df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
dfex.ChestPainType.unique()
```

```python

dfex.RestingECG.unique()
```

```python

dfex.ExerciseAngina.unique()
```

```python
dfex.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
X
```

```python

from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_pca.shape
```

```python
X.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```
