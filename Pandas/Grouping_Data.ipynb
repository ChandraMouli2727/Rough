```python
def grouper(df,idx,col):
     if 80 <= df[col].loc[idx] <=90:
        return '80-90'
     elif 50 <= df[col].loc[idx]<=60:
        return '50-60'
     else:
        return 'others'
```

```python
g = df.groupby(lambda idx : grouper(df,idx,"temperature"))
g
```

```python
g.describe()
```

```python
for key,d in g:
    print("Key : ",key)
    print("\n")
    print("Data :",d)
  
```
