```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
'city' : ['mumbai','delhi','banglore'],
'temperature':[32,45,30],
'humidity':[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
'city' : ['new york','chicago','dallas'],
'temperature':[21,14,35],
'humidity':[68,65,75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df = pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
df = pd.concat([india_weather,us_weather],keys = ["india","us"])
df
```

```python
df.loc["us"] # if u use keys like this it is helpful
```

```python
temperature_df = pd.DataFrame({
'city':['mumbai','delhi','banglore'],
'temperature':[32,45,30]
},index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
'city':['delhi','mumbai'],
'windspeed':[7,12]
},index=[1,0])
windspeed_df
```

```python
pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
#merge
df1 = pd.DataFrame({
'city' : ['new york','chicago','dallas'],
'temperature':[21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
'city' : ['new york','chicago','dallas'],
'humidity':[68,65,75]})
df2
```

```python
df3 = pd.merge(df1,df2, on = 'city')
df3
```

```python
df4 = pd.DataFrame({
'city' : ['new york','chicago','dallas','baltimore'],
'temp':[68,65,75,56]})
df4
```

```python
df5 = pd.DataFrame({
'city' : ['new york','chicago','orlando','san fransico'],
'humidity':[68,65,75,45]})
df5
```

```python
df6 = pd.merge(df4,df5,on='city')
df6   #because it is given common in 2 df so by default it is inner join
```

```python
df6 = pd.merge(df4,df5,on='city',how='left')
df6 
```

```python
df6 = pd.merge(df4,df5,on='city',how='right')
df6 
```

```python
df6 = pd.merge(df4,df5,on='city',how='outer')
df6 
```
