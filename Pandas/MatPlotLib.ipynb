```python
import pandas as pd
df_sales = pd.read_excel("C:\\Users\\Chandra Mouli\\Desktop\\Ai_Eng\\CSV_Files\\linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize = (12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color = 'blue',label = 'Fridge') # X axis and y axis
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color='green',label = 'Dishwasher')
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color='orange',label = 'Washing Machine')
plt.title("Product Sales")
plt.ylabel("Revenue (min $)")
plt.xlabel("Financial Quarter")
plt.legend()
plt.show()
```

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
#Pychart
plt.pie(total_sales,labels=total_sales.index,autopct="%1.2f%%",explode=(0.1,0,0),shadow = True,
               startangle = 140) #labels = ['a','b','c']
plt.show()
```
