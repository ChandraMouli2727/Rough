```python
def standardize_currency(curr):
     if curr == "$$" or curr == "Dollars":
         return "USD"
     return curr
df_finicials = pd.read_excel("C:\\Users\\Chandra Mouli\\Desktop\\Ai_Eng\\CSV_Files\\movies_db.xlsx","financials",converters = {
  'currency':standardize_currency 

})
df_finicials.head(5) #use google to get ideas about conversions
```

```python
df_merged = pd.merge(df_movies,df_finicials,on = "movie_id")
df_merged.head(4)
```
