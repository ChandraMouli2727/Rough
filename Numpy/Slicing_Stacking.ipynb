```python
# Customer Id , Name
c = np.array([
       [101,'Mira'],
       [102,'Abdul'],
       [103,'Andrea']
])
#Customer Id,Purchase amount ,Release Date
d = np.array([
    [101,250.50 ,'2026-03-28'],
    [102,150.00,'2026-03-29'],
    [103,300.75,'2026-03-30']
])
```

```python
#Horizontal Stack
np.hstack((c,d)) #Give like a Tuple
```

```python
c
```

```python
e = np.array([
             [104,'sam'],
             [105,'Ram'],
             [106,'Tom']
])
e
```

```python
#Vertical Stack
np.vstack((c,e))
```

```python
transactions = np.array([
       [101, 'Mira',250.5, '2026-03-28'],
       [102, 'Abdul',150.0, '2026-03-29'],
       [103, 'Andrea',300.75, '2026-03-30'],
       [104,'ram',400,'2026-03-31'],
       [105,'sam',500,'2026-04-01']
       ])
transactions
```

```python
#horizontal split the columns
np.hsplit(transactions,[3])
```

```python
a,b = np.hsplit(transactions,[2])
a
```

```python
b
```

```python
#Vertical split
np.vsplit(transactions,[4])
```

```python

x,y = np.vsplit(transactions,[2])
x
```
