```python
import numpy as np
```

```python
#Sales Data for Quarter 1 (Matrix 1)
# Rows Represent Different Products,Columns represent Diferrent Regions USA India China , Iphone Ipod ,Macbook

q1 = np.array([
      [200,220,250], # Product A
      [150,180,210], # Product B
      [300,330,360]  # Product C
])
q2 = np.array([
      [209,231,259], # Product A
      [155,192,222], # Product B
      [310,340,375]  # Product C
])
```

```python
q = q1 + q2
q
```

```python
q2-q1 # Diff units between quarter
```

```python
# In terms of Percentage
(q2-q1)*100/q1
```

```python
prices = np.array([
   [10,12,11], # Prices for Product A in Different regions
   [8,9,10],   # Prices for Product B
   [15,16,17]
])

q1_rev = q1 * prices
q1_rev
```

```python
 # If you give discount 
q1_discounts = q1_rev * 0.2 # Multiply with scalar value (single value)
q1_discounts
```

```python
net_rev = q1_rev - q1_discounts
net_rev
```

```python
np.sum(q1_discounts) # To check total discounts
```

```python
np.sum(net_rev)
```

```python
q2_rev = q2 * prices
q2_rev
```

```python
# Need To explore dot Product & Croos Product in Matrices

features = np.array([
     [2000,3], # House 1 sq,bed room
     [1800,2]  # House 2
])
weights = np.array([150,50000])
# 2000 * 150 + 3 * 50000
np.dot(features,weights)

```

```python
c = np.array([1,2,3])
d = np.array([4,5,6])
np.cross(c,d)
#(c2d3-c3d2,c1d3-c3d1,c1d2-c2d1)
```

* Using Numpy,One can perform Matrix Operations in a convenient way
 * For two Matrices (m1 and m2), one can do addition,subtraction,division, and multiplication by using +,-,/ and * Operators

 *np.dot(m1,m2) and np.cross(m1,m2) can be used to perform a dot and cross products between two metrices (m1 and m2)
